# LOAD THE DATA

```python
import mne
import os
import numpy as np
from sklearn.preprocessing import StandardScaler

# Directory containing EEG data files
data_dir = 'C:/Users/Asus/Desktop/IIT ROORKEE RESEARCH INTERN/eeg-during-mental-arithmetic-tasks-1.0.0'

# List all data files
data_files = os.listdir(data_dir)
data_files.sort()

# Initialize lists to store data and labels
X_data_rest = []
X_data_task = []
subject_ids = []

# Load and preprocess each subject's data
for file in data_files:
    if file.endswith('.edf'):
        
        raw = mne.io.read_raw_edf(os.path.join(data_dir, file), preload=True)
        
        
        raw.filter(l_freq=0.5, h_freq=40)  # Example bandpass filter
        data = raw.get_data()
        

        scaler = StandardScaler()
        data_normalized = scaler.fit_transform(data)
        
        # Segment data into epochs 
        sfreq = raw.info['sfreq']  
        epochs = mne.make_fixed_length_epochs(raw, duration=2.0, preload=True)  
        
        # Extract subject ID and state (rest or task)
        subject_id = file.split('_')[0]
        state = file.split('_')[1].split('.')[0]
        
        if state == '1':
            X_data_rest.append(epochs.get_data())
        elif state == '2':
            X_data_task.append(epochs.get_data())
        
        subject_ids.append(subject_id)

# Convert lists to numpy arrays
X_data_rest = np.concatenate(X_data_rest, axis=0)
X_data_task = np.concatenate(X_data_task, axis=0)   

print(f"Rest data shape: {X_data_rest.shape}")
print(f"Task data shape: {X_data_task.shape}")
```

# Power Spectral Density (PSD) Analysis

```python
import matplotlib.pyplot as plt
import mne

def calculate_psd_multitaper(data, sfreq):
    psds, freqs = mne.time_frequency.psd_array_multitaper(data, sfreq=sfreq, fmin=0.5, fmax=sfreq/2)
    return psds, freqs

# Defining frequency bands 
freq_bands = {
    'Delta': [1, 4],
    'Theta': [4, 8],
    'Alpha': [8, 12],
    'Beta': [12, 30],
    'Gamma': [30, 100]
}

# Calculate PSD for before task (rest state) and during task (task state) using multitaper
sfreq = raw.info['sfreq'] 

psds_rest, freqs = calculate_psd_multitaper(X_data_rest, sfreq)
psds_task, _ = calculate_psd_multitaper(X_data_task, sfreq)


def plot_psd_comparison(psds1, psds2, freqs, freq_band, band_name):
    plt.figure(figsize=(10, 6))
    plt.plot(freqs, 10 * np.log10(psds1.mean(axis=0).mean(axis=0)), color='b', label='Before Task (Rest)')
    plt.plot(freqs, 10 * np.log10(psds2.mean(axis=0).mean(axis=0)), color='r', label='During Task (Task)')
    plt.xlim(freq_band)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power Spectral Density (dB/Hz)')
    plt.title(f'Power Spectral Density ({band_name} band)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot PSD comparisons 
for band_name, (fmin, fmax) in freq_bands.items():
    plot_psd_comparison(psds_rest, psds_task, freqs, (fmin, fmax), band_name)

```

# EXTRACT RELEVANT FEATURES

```python
# Function to extract band power features from PSD data
def extract_features(psds, freqs, freq_bands):
    features = []
    for psd in psds:
        band_powers = []
        for band_name, (fmin, fmax) in freq_bands.items():
            band_power = psd[:, (freqs >= fmin) & (freqs <= fmax)].mean(axis=1)
            band_powers.append(band_power)
        features.append(np.concatenate(band_powers))
    return np.array(features)

# Extract features for rest and task states
X_features_rest = extract_features(psds_rest, freqs, freq_bands)
X_features_task = extract_features(psds_task, freqs, freq_bands)

# Create labels: 0 for rest, 1 for task
y_labels_rest = np.zeros(X_features_rest.shape[0])
y_labels_task = np.ones(X_features_task.shape[0])

X_features = np.concatenate([X_features_rest, X_features_task], axis=0)
y_labels = np.concatenate([y_labels_rest, y_labels_task], axis=0)

print(f"Features shape: {X_features.shape}")
print(f"Labels shape: {y_labels.shape}")
```

# BINARY CLASSIFICATION

## EEGNET


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, BatchNormalization, Activation, AveragePooling2D, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


def create_eegnet_model(input_shape):
    model = Sequential()

    # Layer 1
    model.add(Conv2D(16, (1, 64), padding='same', input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(AveragePooling2D((1, 4), padding='same'))
    model.add(Dropout(0.5))

    # Layer 2
    model.add(Conv2D(32, (2, 1), padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(AveragePooling2D((1, 4), padding='same'))
    model.add(Dropout(0.5))

    # Layer 3
    model.add(Flatten())
    model.add(Dense(2, activation='softmax'))

    return model

# Reshape data for EEGNet
X_features_reshaped = X_features.reshape(X_features.shape[0], X_features.shape[1], 1, 1)

# Split 
X_train, X_test, y_train, y_test = train_test_split(X_features_reshaped, y_labels, test_size=0.2, random_state=42)

# EEGNet model
input_shape = X_train.shape[1:]
eegnet_model = create_eegnet_model(input_shape)
eegnet_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the EEGNet model
history = eegnet_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

## METRICS OF EEGNet MODEL

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Predict the labels for the test set
y_pred_eegnet = eegnet_model.predict(X_test)
y_pred_eegnet = np.argmax(y_pred_eegnet, axis=1)

# Calculate 
accuracy_eegnet = accuracy_score(y_test, y_pred_eegnet)
precision_eegnet = precision_score(y_test, y_pred_eegnet, average='weighted')
recall_eegnet = recall_score(y_test, y_pred_eegnet, average='weighted')
f1_eegnet = f1_score(y_test, y_pred_eegnet, average='weighted')

# Print 
print("EEGNet Model Metrics")
print(f"Accuracy: {accuracy_eegnet:.4f}")
print(f"Precision: {precision_eegnet:.4f}")
print(f"Recall: {recall_eegnet:.4f}")
print(f"F1 Score: {f1_eegnet:.4f}")

# Confusion matrix and classification report
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_eegnet))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_eegnet))
```

## TSCeption

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization, Activation, concatenate
from sklearn.model_selection import train_test_split

def create_tsception_model(input_shape):
    input_layer = Input(shape=input_shape)
    
    # Temporal block 1
    conv1 = Conv2D(32, (1, 3), padding='same')(input_layer)
    conv1 = BatchNormalization()(conv1)
    conv1 = Activation('relu')(conv1)
    pool1 = MaxPooling2D((1, 2), padding='same')(conv1)
    
    # Temporal block 2
    conv2 = Conv2D(32, (1, 3), padding='same')(pool1)
    conv2 = BatchNormalization()(conv2)
    conv2 = Activation('relu')(conv2)
    pool2 = MaxPooling2D((1, 2), padding='same')(conv2)
    
    # Spatial block
    conv3 = Conv2D(32, (input_shape[0], 1), padding='valid')(pool2)
    conv3 = BatchNormalization()(conv3)
    conv3 = Activation('relu')(conv3)
    pool3 = MaxPooling2D((1, 2), padding='same')(conv3)
    
    flatten = Flatten()(pool3)
    dense = Dense(128, activation='relu')(flatten)
    output_layer = Dense(2, activation='softmax')(dense)
    
    model = Model(inputs=input_layer, outputs=output_layer)
    return model

# Reshape data for TSCeption 
X_features_reshaped = X_features.reshape(X_features.shape[0], X_features.shape[1], 1, 1)

# Split 
X_train, X_test, y_train, y_test = train_test_split(X_features_reshaped, y_labels, test_size=0.2, random_state=42)

# Create TSCeption model
input_shape = X_train.shape[1:]
tsception_model = create_tsception_model(input_shape)
tsception_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the TSCeption model
history_tsception = tsception_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

## METRICS OF TSCeption MODEL

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Predict the labels for the test set
y_pred_tsception = tsception_model.predict(X_test)
y_pred_tsception = np.argmax(y_pred_tsception, axis=1)

# Calculate 
accuracy_tsception = accuracy_score(y_test, y_pred_tsception)
precision_tsception = precision_score(y_test, y_pred_tsception, average='weighted')
recall_tsception = recall_score(y_test, y_pred_tsception, average='weighted')
f1_tsception = f1_score(y_test, y_pred_tsception, average='weighted')

# Print 
print("TSCeption Model Metrics")
print(f"Accuracy: {accuracy_tsception:.4f}")
print(f"Precision: {precision_tsception:.4f}")
print(f"Recall: {recall_tsception:.4f}")
print(f"F1 Score: {f1_tsception:.4f}")

# Confusion matrix and classification report
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_tsception))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_tsception))
```
